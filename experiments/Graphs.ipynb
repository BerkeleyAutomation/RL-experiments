show() runs takes in a variable and range and evaluates both the control and segmented reward for 400 episodes - evaluating performance at every 10 episodes. Another important note is the TERMINAL_REWARD aloong with the PIT_REWARD. 

```python
%%capture

import run_transfer as rt
x_ = [x * 0.05 for x in range (0, 10)]
ctrl, res = rt.show("agent_eps", x_)
```

```python
plt.plot(x_, ctrl, 'k') #black
plt.plot(x_, res)
plt.ylabel("AUC")
plt.yticks(fontsize="10")
plt.xticks(fontsize="10")
plt.title("Agent Eps vs AUC")
```

```python
%%capture

x_2 = [x * 0.05 for x in range (0, 16)]
ctrl_2, res_2 = rt.show("env_noise", x_2)
```
